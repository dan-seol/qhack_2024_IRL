```python
import pyzx as zx
from copy import deepcopy
qubit_amount = 5
gate_count = 80
#Generate random circuit of Clifford gates
circuit = zx.generate.cliffordT(qubit_amount, gate_count)
#If running in Jupyter, draw the circuit
zx.draw(circuit)
```

```python
#Use one of the built-in rewriting strategies to simplify the circuit
circuit_1 = deepcopy(circuit)
zx.simplify.full_reduce(circuit_1)
#See the result
zx.draw(circuit_1)
```

```python
circuit_2 = zx.simplify.teleport_reduce(circuit)
zx.draw(circuit_2)
```
